```python
import matplotlib.pyplot as plt

monthly_sales = df.groupby("month_number")["sales"].sum()

plt.figure(figsize=(10,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(range(1,13))
plt.grid(True)

plt.show()


'''**Observation:**
- December recorded the highest sales.
- February recorded the lowest sales.

**Business Insight:**
Sales peak during December, likely due to year-end holiday shopping and promotional campaigns.
    The company should increase inventory, staffing, and marketing efforts before December. 
    Since February has the lowest sales, targeted promotions and discounts could help improve performance during this period.'''
```

```python
import duckdb
import pandas as pd
```

```python

con = duckdb.connect()
con.register("sales", df)
```

```python
con.sql("""
SELECT COUNT(*) AS total_rows
FROM sales
""").df()
```

```python
con.sql("""
SELECT
    ROUND(SUM(sales), 2) AS total_sales,
    ROUND(SUM(profit), 2) AS total_profit,
    COUNT(DISTINCT order_id) AS total_orders
FROM sales;
""").df()
```

```python
con.sql("""
SELECT
    country,
    ROUND(SUM(sales),2) AS total_sales,
    ROUND(SUM(profit),2) AS total_profit
FROM sales
GROUP BY country
ORDER BY total_sales DESC
LIMIT 10;
""").df()
```

```python
con.sql("""
SELECT
    customer_name,
    ROUND(SUM(sales), 2) AS total_sales,
    ROUND(SUM(profit), 2) AS total_profit
FROM sales
GROUP BY customer_name
ORDER BY total_sales DESC
LIMIT 10;
""").df()
```

```python
con.sql("""
SELECT
    category,
    ROUND(SUM(sales),2) AS total_sales,
    ROUND(SUM(profit),2) AS total_profit
FROM sales
GROUP BY category
ORDER BY total_sales DESC;
""").df()
```

```python
con.sql("""
SELECT
    product_name,
    ROUND(SUM(sales), 2) AS total_sales,
    ROUND(SUM(profit), 2) AS total_profit
FROM sales
GROUP BY product_name
ORDER BY total_sales DESC
LIMIT 10;
""").df()
```

```python
con.sql("""
SELECT
    month,
    ROUND(SUM(sales),2) AS total_sales,
    ROUND(SUM(profit),2) AS total_profit
FROM sales
GROUP BY month
ORDER BY MIN(month_number);
""").df()
```

```python
con.sql("""
SELECT
    market,
    ROUND(SUM(sales),2) AS total_sales,
    ROUND(SUM(profit),2) AS total_profit
FROM sales
GROUP BY market
ORDER BY total_sales DESC;
""").df()
```

```python
con.sql("""
SELECT
    ship_mode,
    COUNT(*) AS total_orders,
    ROUND(AVG(shipping_days),2) AS avg_shipping_days,
    ROUND(SUM(sales),2) AS total_sales
FROM sales
GROUP BY ship_mode
ORDER BY total_sales DESC;
""").df()
```

```python
con.sql("""
SELECT
    state,
    ROUND(SUM(sales),2) AS total_sales,
    ROUND(SUM(profit),2) AS total_profit
FROM sales
GROUP BY state
ORDER BY total_sales DESC
LIMIT 10;
""").df()
```

```python
con.sql("""
SELECT
    product_name,
    ROUND(SUM(profit),2) AS total_profit,
    ROUND(SUM(sales),2) AS total_sales
FROM sales
GROUP BY product_name
ORDER BY total_profit DESC
LIMIT 10;
""").df()
```

```python
con.sql("""
SELECT
    product_name,
    ROUND(SUM(profit),2) AS total_profit,
    ROUND(SUM(sales),2) AS total_sales
FROM sales
GROUP BY product_name
ORDER BY total_profit ASC
LIMIT 10;
""").df()
```

```python
con.sql("""
SELECT
    category,
    ROUND(AVG(profit_margin),2) AS avg_profit_margin,
    ROUND(SUM(sales),2) AS total_sales,
    ROUND(SUM(profit),2) AS total_profit
FROM sales
GROUP BY category
ORDER BY avg_profit_margin DESC;
""").df()
```

```python
con.sql("""
SELECT
    product_name,
    ROUND(SUM(sales),2) AS total_sales,
    RANK() OVER (ORDER BY SUM(sales) DESC) AS sales_rank
FROM sales
GROUP BY product_name
LIMIT 10;
""").df()
```

```python
con.sql("""
WITH ranked_products AS (
    SELECT
        category,
        product_name,
        ROUND(SUM(sales),2) AS total_sales,
        ROW_NUMBER() OVER (
            PARTITION BY category
            ORDER BY SUM(sales) DESC
        ) AS product_rank
    FROM sales
    GROUP BY category, product_name
)
SELECT
    category,
    product_name,
    total_sales,
    product_rank
FROM ranked_products
WHERE product_rank <= 3
ORDER BY category, product_rank;
""").df()
```

```python
con.sql("""
SELECT
    customer_name,
    ROUND(SUM(profit),2) AS total_profit,
    DENSE_RANK() OVER (
        ORDER BY SUM(profit) DESC
    ) AS profit_rank
FROM sales
GROUP BY customer_name
LIMIT 5;
""").df()
```

```python
con.sql("""
SELECT
    month_number,
    month,
    ROUND(SUM(sales),2) AS monthly_sales,
    ROUND(
        SUM(SUM(sales)) OVER (
            ORDER BY month_number
        ),2
    ) AS running_total
FROM sales
GROUP BY month_number, month
ORDER BY month_number;
""").df()
```

```python
con.sql("""
SELECT
    CASE
        WHEN total_sales >= 10000 THEN 'High Value'
        WHEN total_sales >= 5000 THEN 'Medium Value'
        ELSE 'Low Value'
    END AS customer_segment,
    COUNT(*) AS customer_count,
    ROUND(AVG(total_sales),2) AS avg_sales
FROM (
    SELECT
        customer_name,
        SUM(sales) AS total_sales
    FROM sales
    GROUP BY customer_name
) t
GROUP BY customer_segment
ORDER BY avg_sales DESC;
""").df()
```

```python
con.sql("""
SELECT
    year,
    ROUND(SUM(sales),2) AS total_sales,
    ROUND(SUM(profit),2) AS total_profit,
    COUNT(DISTINCT order_id) AS total_orders
FROM sales
GROUP BY year
ORDER BY year;
""").df()
```

```python
con.sql("""
WITH ranked_products AS (
    SELECT
        category,
        product_name,
        ROUND(SUM(profit),2) AS total_profit,
        RANK() OVER (
            PARTITION BY category
            ORDER BY SUM(profit) DESC
        ) AS profit_rank
    FROM sales
    GROUP BY category, product_name
)
SELECT
    category,
    product_name,
    total_profit,
    profit_rank
FROM ranked_products
WHERE profit_rank <= 5
ORDER BY category, profit_rank;
""").df()

```

```python
con.sql("""
WITH yearly_sales AS (
    SELECT
        year,
        SUM(sales) AS total_sales
    FROM sales
    GROUP BY year
)
SELECT
    year,
    ROUND(total_sales,2) AS total_sales,
    ROUND(
        LAG(total_sales) OVER (ORDER BY year),
        2
    ) AS previous_year_sales,
    ROUND(
        ((total_sales - LAG(total_sales) OVER (ORDER BY year))
        / LAG(total_sales) OVER (ORDER BY year)) * 100,
        2
    ) AS yoy_growth_percent
FROM yearly_sales
ORDER BY year;
""").df()
```
